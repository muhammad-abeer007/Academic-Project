# Cleaning and Checking Assciation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
sns.set()
```

```python
df = pd.read_csv('street_children_mod2.csv')
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df.shape
```

# Remove Missing values if exist 

```python
value_to_remove = 88 or 99
mask = (df == value_to_remove).any(axis=1)
df = df[~mask]
```

```python
value_to_remove1 = '88' or '99'
mask1 = (df == value_to_remove1).any(axis=1)
df = df[~mask1]
```

```python
df.shape
```

# Checking If any category contains less than 5%

```python
threshold = 0.05  # 5% threshold
list_freq = []

for column in df.columns:
    if df[column].dtype == 'object':
        value_counts = df[column].value_counts(normalize=True)
        infrequent_categories = value_counts[value_counts < threshold]
        if not infrequent_categories.empty:
            list_freq.append(column)
```

```python
list_freq
```

```python
df = df.drop(list_freq, axis=1)
```

```python
df
```

```python
df.shape
```

```python
df.columns.values
```

# Checking the association with 'Arrested_by_police'

```python
fixed_var1 = 'Arrested_by_police'
list_sig1 = []


# Perform the chi-square test of independence for each variable against the fixed variable
for variable in df.columns:
    if variable != fixed_var1:
        contingency_table = pd.crosstab(df[fixed_var1], df[variable])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        '''print("Chi Square Value: ", chi2)
        print("Degree of Freedm : ", dof)
        print("p-value : ", expected)
        print(" Expected Frequency :", expected)'''
       
        if p<.06:
            list_sig1.append(variable)
        
```

```python
list_sig1
```
